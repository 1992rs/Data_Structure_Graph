```python
# Minimum number of squares
import sys 
import math 

def msp(n , dp):
  if n == 0:
    return 0
  ans = sys.maxsize

  root = int(math.sqrt(n))
  for i in range(1 , root + 1):
    if dp[n-(i**2)] == -1:
      minans = 1 + msp((n-(i**2)), dp)
      dp[n-(i**2)] = minans
    else:
      minans = dp[n-(i**2)]
    ans = min(ans , minans) 
  return ans 

n = int(input())
dp = [-1 for i in range(n+1)]
print(msp(n , dp))
```

```python
# DP-1
import sys
def minstepsto1dp(n):
  dp = [-1 for i in range(n+1)]
  dp[0] = 0
  dp[1] = 0
  for i in range(2 , n+1):
    op1 = dp[i-1]
    if i % 2 == 0:
      op2 = dp[i//2]
    else:
      op2 = sys.maxsize
    
    if i % 3 == 0:
      op3 = dp[i//3]
    else:
      op3 = sys.maxsize
    dp[i] = 1 + min(op1 , op2 , op3)
  return dp[n]

n = int(input())
print(minstepsto1dp(n))
```

```python
def longestPalindrome(s):
      dp = [[False for i in range(len(s))] for i in range(len(s))]
      for i in range(len(s)):
         dp[i][i] = True
      max_length = 1
      start = 0
      for l in range(2,len(s)+1):
         for i in range(len(s)-l+1):
            end = i+l
            if l==2:
               if s[i] == s[end-1]:
                  dp[i][end-1]=True
                  max_length = l
                  start = i
            else:
               if s[i] == s[end-1] and dp[i+1][end-2]:
                  dp[i][end-1]=True
                  max_length = l
                  start = i
      return s[start:start+max_length]
ans = longestPalindrome("ABBABBC")
print('result:', ans)
```

```python
# GRAPH
## Read input as specified in the question.
## Print output as specified in the question.
c = 0


class Graph:
    def __init__(self, nVertices):
        self.nVertices = nVertices
        self.adjMatrix = [[0 for i in range(nVertices)] for j in range(nVertices)]

    def addEdge(self, v1, v2):  # Take v1 and v2 and add the edge between them
        self.adjMatrix[v1][v2] = 1
        self.adjMatrix[v2][v1] = 1

    def __hasPathHelper(self, visited, v1, v2):
        global c
        visited[v1] = True
        for i in range(self.nVertices):
            if self.adjMatrix[v1][i] > 0 and visited[i] is False:
                self.__hasPathHelper(visited, i, v2)
                if i == v2:
                    c = 1
        return c

    def hasPath(self, v1, v2):
        visited = [False for i in range(self.nVertices)]
        p = self.__hasPathHelper(visited, v1, v2)
        return p


li = [int(x) for x in input().split()]
v = li[0]
e = li[1]
g = Graph(v)
for i in range(e):
    a, b = [int(x) for x in input().split()]
    g.addEdge(a,b)
li1 = [int(x) for x in input().split()]
v1 = li1[0]
v2 = li1[1]
result = g.hasPath(v1,v2)
if result == 1:
  print('true')
else:
  print('false')
```
